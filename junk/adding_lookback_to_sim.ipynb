```python
sim = Sim()
```

```python
stock_symbols = ['NVDA', 'MSFT']
interval_inputs = {'1m': {'columns': ['open', 'high', 'close', 'volume'], 'lookback': 100}, '1d': {'columns': ['open', 'high', 'close', 'volume'], 'lookback': 100}}
start_date = pd.Timestamp(datetime(2024,3,15)).tz_localize('UTC').tz_convert('EST')
end_date = pd.Timestamp(datetime(2024,9,18)).tz_localize('UTC').tz_convert('EST')
lookback = {'1m': 100, '1d': 100}
```

```python
combined_df = sim.data.update_price_data(stock_symbols, interval_inputs, data_folder='/Users/vandanchopra/Vandan_Personal_Folder/CODE_STUFF/Projects/mathematricks/db/data/yahoo', throttle_secs=0.25, back_test_start_date=start_date, back_test_end_date=end_date, lookback=lookback)
```

```python
joined_dict = {}
for interval, lookback_value in lookback.items():
    before = combined_df.loc[interval].loc[:start_date]
    after = combined_df.loc[interval].loc[start_date:end_date]
    # get the last 20 rows of before
    before_new = before.iloc[-lookback_value:]
    print({'interval': interval, 'before':before.shape, 'before_new': before_new.shape, 'after': after.shape})
    # join before and after dataframes
    joined = pd.concat([before_new, after])
    joined.sort_index(inplace=True)
    joined_dict[interval] = joined
combined_df = pd.concat(joined_dict.values(), keys=joined_dict.keys(), names=['interval', 'date'])
```
